```python
from zipfile import ZipFile
import pickle
import matplotlib.pyplot as plt
```

```python
zip = ZipFile('DIEF_B_Snippet50_3weeks.pkl.zip', 'r')
l_pkl = pickle.loads(zip.read('DIEF_B_Snippet50_3weeks.pkl'))
```

```python
plt.figure(figsize=[15,8])
plt.title('DIEF_B_Snippet50_3weeks.pkl.zip')
for ii, i_dict in enumerate(l_pkl):
    plt.scatter(i_dict['t'], i_dict['v'], alpha=.01)
plt.axis((18610, 19003, -10, 90))
```

```python
# plot the last n_streams
n_streams = 20
for ii, i_dict in enumerate(l_pkl[-n_streams:]):
    plt.figure(figsize=(10, 3))
    plt.title(i_dict['y'] + ' len=' + str(len(i_dict['t'])) + '\n' + i_dict['StreamID'])
    plt.scatter(i_dict['t'], i_dict['v'])
```

```python
plt.figure(figsize=[15,8])
plt.title('DIEF_B_Snippet50_3weeks.pkl.zip: Histograms of the values.')
for ii, i_dict in enumerate(l_pkl):
    plt.hist(i_dict['v'], bins=100, histtype='step', alpha=.2, log=True)
plt.axis([-1, 101, 0, 40000])
```
